#  Find out Top &amp; least sold products

```python
rating_counts =df.groupby(['product'])['rating'].count().reset_index()
top_sold = rating_counts.sort_values(by='rating', ascending=False).head(5)

# Least 5 sold products
least_sold = rating_counts.sort_values(by='rating', ascending=True).head(5)

top_sold.head(), least_sold.head()
```

# Measuring discount on a certain item.

```python
df['discount']=df['market_price']-df['sale_price']
```

```python
df['discount']=(df['discount']/df['market_price'])*100
```

```python
df.head(5)
```

#  Find out the Missing Values from the Dataset

```python
 df.isnull().sum()
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mean())
df['description'] = df['description'].fillna('Unknown')
df['discount'] = df['discount'].fillna(df['discount'].mean())
df.isnull().sum()
```

# Find out the outliers from the dataset according to the columns and fill them with the mean

```python
columns = ['sale_price', 'market_price', 'rating']

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Clip outliers
    df[col] = df[col].clip(lower, upper)

sns.boxplot(x=df['sale_price'])
plt.title('Sale_price(Box Plot)')
plt.show
```

# VISUALIZATIONS

```python

# count values
counts = df['category'].value_counts()

# bar plot

plt.bar(counts.index, counts.values)

# labels & title
plt.xticks(rotation=45, ha='right')
plt.ylabel('Count')
plt.title('Category-wise Count')

plt.show()
```

# Top 5 Brands Distribution

```python
top5_brands = df['brand'].value_counts().head(5).index
df_top5 = df[df['brand'].isin(top5_brands)]

plt.figure(figsize=(6,4))
sns.histplot(data=df_top5, x='brand')

plt.title('Top 5 Brands Distribution')
plt.xlabel('Brand')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Market Price Vs Sales Price

```python

sns.scatterplot(data=df, y='market_price', x='sale_price')

plt.title('Market Price vs Sales Price')
plt.tight_layout()
plt.show()
```

# Rating

```python
top10_rating = df['rating'].value_counts().head(15).index
df_top10 = df[df['rating'].isin(top10_rating)]

sns.histplot(x='rating', data=df_top10,bins = 10,shrink = 0.8)

plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Rating Distribution')

plt.tight_layout()
plt.show()
```

# Top 7 Category Distribution

```python
import matplotlib.pyplot as plt

# count categories
category_counts = df['category'].value_counts().head(7)

plt.figure(figsize=(6,6))
plt.pie(
    category_counts.values,
    labels=category_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    labeldistance=1.2,   # ðŸ‘ˆ move labels out
    pctdistance=0.7  
)

plt.title(' Top 7 Category Distribution')
plt.tight_layout()
plt.show()
```
